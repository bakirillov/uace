```python
import os
import numpy as np
import pandas as pd
import os.path as op
```

```python
singles = [a for a in os.walk("../results/on-target")][0][2]
data = pd.concat(
    [pd.read_csv(op.join("../results/on-target", a), index_col=0) for a in singles]
)
data.index = [" ".join(a.split("_")[2:]).replace(".csv", "") for a in singles]
data = data.sort_values(by="SCC", ascending=False)
```
